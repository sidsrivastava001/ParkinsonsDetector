```python
patient_dict = {}
with open(r"C:\Users\dkuma\Downloads\gait-in-parkinsons-disease-1.0.0\demographics.txt", 'r') as file:
    data = file.read().splitlines()
    data.pop(0)
    for line in data:
        linedata = line.split()
        if(linedata[8] == "NaN"):
            patient_dict[linedata[0]] = 0.0
        else:
            patient_dict[linedata[0]] = float(linedata[8])
patient_dict
```

```python
import keras
from keras.layers import Dense, Flatten, Dropout
from keras.layers import Conv1D, GlobalAveragePooling2D, MaxPooling1D, Activation
from keras.models import Sequential
from keras.callbacks import History 
history = History()


input_shape = (4034, 18)

num_classes = 2
batch_size = 5
epochs = 40

 
Y_train = keras.utils.to_categorical(Y_train, num_classes)
y_test = keras.utils.to_categorical(y_test, num_classes)

print(Y_train)
print(y_test)

model = Sequential()
model.add(Conv1D(filters=16, kernel_size=3, activation='relu', input_shape=input_shape))
model.add(MaxPooling1D(pool_size=2))
model.add(Dropout(0.4))

model.add(Conv1D(filters=16, kernel_size=3, activation='relu'))
model.add(MaxPooling1D(pool_size=2))
model.add(Dropout(0.4))

model.add(Conv1D(filters=32, kernel_size=3, activation='relu'))
model.add(MaxPooling1D(pool_size=2))
model.add(Dropout(0.4))

model.add(Flatten())
model.add(Dense(100, activation='relu'))
model.add(Dense(20, activation='relu'))
model.add(Dense(2, activation='softmax'))
opt = keras.optimizers.RMSprop(learning_rate=0.0001)
model.compile(loss='binary_crossentropy', optimizer=opt, metrics=['accuracy'])

model.summary()

callbacks = [history]

model.fit(X_train, Y_train, batch_size=batch_size, epochs=epochs,
          validation_data=(x_test, y_test), callbacks=callbacks, verbose=1)
 
train_score = model.evaluate(X_train, Y_train, verbose=0)
print('Train loss: {}, Train accuracy: {}'.format(train_score[0], train_score[1]))
test_score = model.evaluate(x_test, y_test, verbose=0)
print('Test loss: {}, Test accuracy: {}'.format(test_score[0], test_score[1]))
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from keras.callbacks import History 
history = History()

first_conv = [None]*10
first_conv_output = [None]*10

num_classes = 2
batch_size = 20
epochs = 25

 
Y_train = keras.utils.to_categorical(Y_train, num_classes)
y_test = keras.utils.to_categorical(y_test, num_classes)

input_shape = (4034,1)

for i in range(0, 10):
    first_conv[i] = keras.Input(shape=input_shape)
    first_conv_output[i] = layers.Conv1D(filters=8, kernel_size=3, activation="relu")(first_conv[i])
    first_conv_output[i] = layers.Conv1D(filters=16, kernel_size=3, activation="relu")(first_conv_output[i])
    first_conv_output[i] = layers.MaxPooling1D(pool_size=2)(first_conv_output[i])
    
    first_conv_output[i] = layers.Conv1D(filters=16, kernel_size=3, activation="relu")(first_conv_output[i])
    first_conv_output[i] = layers.Conv1D(filters=16, kernel_size=3, activation="relu")(first_conv_output[i])
    first_conv_output[i] = layers.MaxPooling1D(pool_size=2)(first_conv_output[i])
    
    first_conv_output[i] = layers.Flatten()(first_conv_output[i])
    first_conv_output[i] = layers.Dense(100, activation="relu")(first_conv_output[i])
    first_conv_output[i] = layers.Dropout(0.5)(first_conv_output[i])
    
    
final_layer = layers.concatenate(first_conv_output)
final_layer = layers.Dropout(0.5)(final_layer)
final_layer = layers.Dense(100, activation="relu")(final_layer)
final_layer = layers.Dropout(0.5)(final_layer)
final_layer = layers.Dense(20, activation="relu")(final_layer)
final_layer = layers.Dropout(0.5)(final_layer)
final_layer = layers.Dense(num_classes, activation="softmax")(final_layer)

model = keras.Model(inputs=first_conv, outputs=final_layer)
model.summary()
callbacks = [history]

model.compile(optimizer=keras.optimizers.RMSprop(1e-3), loss=keras.losses.BinaryCrossentropy(), metrics=['accuracy'])
model.fit([X_train[:,:,0], X_train[:,:,1], X_train[:,:,2], X_train[:,:,3], X_train[:,:,4], X_train[:,:,5], X_train[:,:,6], X_train[:,:,7], X_train[:,:,8], X_train[:,:,9]], Y_train,batch_size=batch_size, epochs=epochs,
         validation_data=([x_test[:,:,0], x_test[:,:,1], x_test[:,:,2], x_test[:,:,3], x_test[:,:,4], x_test[:,:,5], x_test[:,:,6], x_test[:,:,7], x_test[:,:,8], x_test[:,:,9]], y_test),callbacks=callbacks, verbose=1)
```
